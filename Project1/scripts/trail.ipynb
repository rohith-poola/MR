```python
import pybullet as p
import pybullet_data
import time
import numpy as np
import matplotlib.pyplot as plt

# Connect to GUI
p.connect(p.DIRECT)
p.setAdditionalSearchPath(pybullet_data.getDataPath())

# Load ground and Husky robot
p.loadURDF("plane.urdf")
husky_id = p.loadURDF("husky/husky.urdf", [0, 0, 0.2])

# Set gravity
p.setGravity(0, 0, -9.8)

# Identify wheel joints (manually confirmed from URDF or inspection)
wheel_joints = [2, 3, 4, 5]  # front_left, front_right, rear_left, rear_right

positions = []

def set_wheel_velocity(left_vel, right_vel, force=100):
    """Applies velocity to Husky's wheels."""
    for i, joint in enumerate(wheel_joints):
        if i % 2 == 0:  # left wheels
            p.setJointMotorControl2(husky_id, joint, p.VELOCITY_CONTROL, targetVelocity=left_vel, force=force)
        else:  # right wheels
            p.setJointMotorControl2(husky_id, joint, p.VELOCITY_CONTROL, targetVelocity=right_vel, force=force)

def move_husky(left_vel, right_vel, duration):
    """Moves Husky with given wheel velocities for `duration` seconds."""
    set_wheel_velocity(left_vel, right_vel)
    for _ in range(duration * 240):
        p.stepSimulation()
        pos, _ = p.getBasePositionAndOrientation(husky_id)
        positions.append(pos[:2])
        time.sleep(1/10000)

# 1. Move straight for 5 sec
move_husky(5, 2, 10)
move_husky(2, 5, 10)

# Disconnect
p.disconnect()

# Plot trajectory
x_vals, y_vals = zip(*positions)
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, marker='o', linewidth=2)
plt.title("Husky Trajectory (X-Y)")
plt.xlabel("X Position (m)")
plt.ylabel("Y Position (m)")
plt.grid(True)
plt.axis("equal")
plt.show()
```

## Control Sequences

```python
import numpy as np

# Define nominal control sequence for 7 joints and 5 timesteps
nominal_control = np.array([
    [0.0, 0.1, 0.2, -0.1, 0.0, 0.05, -0.05],
    [0.1, 0.2, 0.1, -0.2, 0.1, 0.1, -0.1],
    [0.2, 0.3, 0.0, -0.3, 0.2, 0.15, -0.15],
    [0.3, 0.4, -0.1, -0.4, 0.3, 0.2, -0.2],
    [0.4, 0.5, -0.2, -0.5, 0.4, 0.25, -0.25]
])

# Set number of samples
num_samples = 10

# Define Gaussian noise parameters
noise_mean = 0.0
noise_std = 0.75  # Standard deviation of noise

# Generate samples around the nominal control sequence
samples = np.zeros((num_samples, 5, 7))

for i in range(num_samples):
    noise = np.random.normal(noise_mean, noise_std, nominal_control.shape)
    samples[i] = nominal_control + noise

# Print generated samples
for i in range(num_samples):
    print(f"Sample {i+1}:\n{samples[i]}\n")
```
